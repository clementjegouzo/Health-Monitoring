```python
import numpy as np
import matplotlib.pyplot as plt

import keras
from keras.models import Sequential
from keras.layers import Dense, Softmax

import csv
from sklearn.model_selection import train_test_split
```

```python
path = "Database/Frequences.csv"
fichier = list(csv.reader(open(path,'r'),delimiter=';'))

dims = [len(fichier), len(fichier[0])]

header = fichier[0]
X = []
Y = []
Y_type = []
for i in range(len(fichier[1:])) :
    for j in range(dims[1]) :
        X.append([i+1, float(fichier[i+1][j])])
        if str.lower(header[j][0]) == "s" :
            Y.append(1)
            Y_type.append(np.eye(3)[0,:])
        elif str.lower(header[j][0]) == "h" :
            Y.append(0)
            Y_type.append(np.eye(3)[1,:])
        elif str.lower(header[j][0]) == "t" :
            Y.append(0)
            Y_type.append(np.eye(3)[2,:])

X = np.array(X)
Y = np.array(Y)
Y_type = np.array(Y_type)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
Y_type_train, Y_type_test = train_test_split(Y_type, test_size=0.3)
```

```python
perceptron = Sequential(name="Detection_cassure")
perceptron.add(Dense(1, activation="sigmoid", input_shape=(None,2)))
perceptron.summary()
perceptron.compile(optimizer="Adam",loss=keras.losses.BinaryCrossentropy(), metrics=["acc"])
```

```python
output = perceptron.fit(X_train, Y_train, epochs=1000, validation_data=(X_test, Y_test))
```

```python
plt.plot(output.epoch, output.history['loss'], label='train')
plt.plot(output.epoch, output.history['val_loss'], label='test')
plt.title('Training and validation loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend()
plt.grid()
```

```python
plt.plot(output.epoch, output.history['acc'], label='train')
plt.plot(output.epoch, output.history['val_acc'], label='test')
plt.title('Training and validation accuracy')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend()
plt.grid()
```

```python
perceptron_type = Sequential(name="Detection_type_cassure")
perceptron_type.add(Dense(3, activation="relu", input_shape=(None,2)))
perceptron_type.add(Softmax())
perceptron_type.summary()
perceptron_type.compile(optimizer="Adam",loss=keras.losses.CategoricalCrossentropy(), metrics=["acc"])
```

```python
output_type = perceptron_type.fit(X_train, Y_type_train, epochs=1000, validation_data=(X_test, Y_type_test))
```

```python
plt.plot(output_type.epoch, output_type.history['loss'], label='train')
plt.plot(output_type.epoch, output_type.history['val_loss'], label='test')
plt.title('Training and validation loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend()
plt.grid()
```

```python
plt.plot(output_type.epoch, output_type.history['acc'], label='train')
plt.plot(output_type.epoch, output_type.history['val_acc'], label='test')
plt.title('Training and validation accuracy')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend()
plt.grid()
```
